### 2. theory
1. Calculate the atmospheric refraction value for each date using the formulas of spherical trigonometry. A flat approximation can also be used, but then it will be more difficult to calculate the error of the result.
2. What kind of dependence should atmospheric refraction at the horizon have on pressure and temperature? You can start [from here (formula 3.13)](https://ru.wikipedia.org/wiki/%D0%9F%D0%BE%D0%BA%D0%B0%D0%B7%D0%B0%D1%82%D0%B5%D0%BB%D1%8C_%D0%BF%D1%80%D0%B5%D0%BB%D0%BE%D0%BC%D0%BB%D0%B5%D0%BD%D0%B8%D1%8F#%D0%9F%D0%BB%D0%BE%D1%82%D0%BD%D0%BE%D1%81%D1%82%D1%8C):
   $$\dfrac{n-1}{\rho} = \mathrm{const}$$

   

$$n = C\rho+1$$

$$pV = \frac{M}{\mu}RT$$

$$p = \frac{\rho}{\mu}RT$$

$$\rho = \frac{p\mu}{RT}$$
$$\rho \sim \frac{p}{T}$$

According to Snellius' law:
$$\frac{\sin(\alpha)}{\sin(\beta)} = \frac{n_2}{n_1}$$

We will count the angles from the plane:
$$\frac{\cos(\alpha)}{\cos(\beta)} = \frac{n_2}{n_1}$$

$$
\frac{1-\alpha^2/2}{1-\beta^2/2} = \frac{n_2}{n_1} = \frac{C\rho_2+1}{1} = C\rho_2+1
$$

Than $r = \beta-\alpha$ (this is the value of refraction), some long calculations are obtained there, but we can consider the extreme case (full internal reflection): $\alpha = 0, \rho = \beta$

$$
(1-r^2/2)^{-1} = C\rho+1
$$
$$
(1+r^2/2) \approx C\rho+1
$$
$$r\sim \sqrt{\rho}$$

light passes sequentially through layers with slightly different $\rho_i$, in each of which the situation is close to the option we are considering. The main problem is the non-isothermal nature of the atmosphere, which is especially great at dawn.
