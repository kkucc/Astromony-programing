```python
rd = 180/np.pi
lam = -data['Days_equinox'].values*360/365.25/rd
eps = 23.5/rd
delta = np.arcsin(np.sin(eps)*np.sin(lam))
delta*rd
phi = 80/rd
rho = (35+16)/60/rd
t = np.arccos(-(rho)/(np.cos(phi)*np.cos(delta)) - np.tan(phi)*np.tan(delta))
dt_model = (2*t*rd/15)[1:]-(2*t*rd/15)[:-1]
dt_model
# t*rd/15*2
```
